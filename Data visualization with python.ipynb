```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
x = [1,2,3]
y = [4,6,9]
z = [10,15,0]
plt.plot(x,y)
plt.plot(x,z)
plt.title("test plot")
plt.xlabel("X")
plt.ylabel("Y and Z")
plt.legend(["This is y","This is z"])
plt.show()
```
